### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(df, new_col_name, existing_col_name):
   #df["legal_drinker"] = np.where(df["age"]>17, True, False)  
    df[new_col_name] = np.where(df[existing_col_name]>17, True, False)  
```

```python
majority(dfs, "legal_drinker", "age")
```

```python
dfs.head()
```

```python
dfs[["age", "legal_drinker"]]
```

```python
# an other way
def majority(x):
    if x > 17:
        return True
    else:
        return False
    
dfs['legal_drinker'] = dfs['age'].apply(majority)
dfs.head()
```
